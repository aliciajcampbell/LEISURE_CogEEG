```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme('notebook')
from scipy import stats

df = pd.read_excel('LEISURE_SART_T1_T2_T3_CONCAT.xlsx')
```

```python
df.describe()
```

```python
# Calculating Hit Rate and False Alarm Rate
df['hr'] = df['correct_go'] / df['total_go']
df['far'] = df['incorrect_no_go'] / df['total_no_go']

# Adjusting for Extreme Values with Log-linear Correction (Half-item Correction)
df['adj_hr'] = (df['correct_go'] + 0.5) / (df['total_go'] + 1)
df['adj_far'] = (df['incorrect_no_go'] + 0.5) / (df['total_no_go'] + 1)

# Z Transformation
df['z_adj_hr'] = stats.norm.ppf(df['adj_hr'])
df['z_adj_far'] = stats.norm.ppf(df['adj_far'])

# Calculate d'
df['d_prime'] = df['z_adj_hr'] - df['z_adj_far']

# Calculate criterion
df['criterion'] = -0.5 * (df['z_adj_hr'] + df['z_adj_far'])
```

```python
df.to_excel("LEISURE_SART_T1_T2_T3_CONCAT.xlsx", index=False)
```
